```python
import requests
from bs4 import BeautifulSoup

# Website ka URL (Replace karein actual URL se)
url = "https://knowmyingredient.com/"  # Actual page ka link dalen

# Website se data fetch karna
response = requests.get(url)

# Agar request successful hai toh process karein
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    # Ingredients ke liye specific HTML tags ya classes ko target karein
    ingredients = soup.find_all('div', class_='ingredient-class')  # Replace 'ingredient-class' with actual class

    # Ingredients ko print karein
    for ingredient in ingredients:
        print(ingredient.text.strip())

else:
    print("Failed to retrieve the webpage. Status Code:", response.status_code)
```
