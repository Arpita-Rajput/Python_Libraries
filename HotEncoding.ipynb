```python
import pandas as pd 

df = pd.DataFrame({"Size":["s" , "m" , "l" , "xl" , "s" , "m" , "l" , "s" , "s" , "l" , "xl" , "m"]})
df.head(3)
```

```python
ord_data  = [["s" , "m" , "l","xl"]] #We use double bracktes because we use double dimension data 
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories = ord_data)
oe.fit(df[["Size"]])
```

```python
df["Size_en"] = oe.transform(df[["Size"]])
df
```

```python
#Through Map function perform encoding 

ord_data1 = {"s" : 0 , "m":1 , "l" : 2 , "xl":3} #we can take any random number 
df["Size_en_map"] = df["Size"].map(ord_data1)
df
```
